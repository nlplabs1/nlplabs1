```python
# import Abiy Ahmed's tweets
aa = pd.read_csv('/content/drive/MyDrive/datasets/aaatweets.csv')
#import the conversation following Abiy Ahmed's tweets
df = pd.read_csv('/content/drive/MyDrive/datasets/aa_audience.csv')
```

```python
# list the tweet language types
aalang = pd.DataFrame(aa.lang.unique())
```

```python
# subset the english language tweets 
aa_en = aa[aa['lang']=='en']
```

```python
# subset the language types other than en for translation to en
aa_amh = aa[aa['lang']=='am']
aa_ar =  aa[aa['lang']=='ar']
aa_eu = aa[aa['lang']=='eu']
aa_in =  aa[aa['lang']=='in']
aa_zxx = aa[aa['lang']=='zxx']
aa_et =  aa[aa['lang']=='et']
aa_und = aa[aa['lang']=='und']
aa_tr =  aa[aa['lang']=='tr']
aa_hi = aa[aa['lang']=='hi']
aa_cy =  aa[aa['lang']=='cy']
aa_qme = aa[aa['lang']=='qme']
aa_vi =  aa[aa['lang']=='vi']
aa_fi =  aa[aa['lang']=='fi']
```

```python
#translate am to en
aa_amh['aa_text'] = aa_amh['text'].apply(lambda x: ts.google(x, from_language='am', to_language='en'))
```

```python
#translate ar to en
aa_ar['aa_text'] = aa_ar['text'].apply(lambda x: ts.google(x, from_language='ar', to_language='en'))
```

```python
aa_eu['aa_text'] = aa_eu['text'].apply(lambda x: ts.google(x, from_language='eu', to_language='en'))
```

```python
aa_et['aa_text'] = aa_et['text'].apply(lambda x: ts.google(x, from_language='et', to_language='en'))
```

```python
# copy the language column to match with the other dfs
aa_en['aa_text'] = aa_en['text']
```

```python
# Merge the dfs
aa_final = pd.concat([aa_en, aa_amh, aa_ar])
```

```python
aa_final.to_csv('/content/drive/MyDrive/datasets/abiy_ahmed_tweets/aa_tweets_translation.csv')
```
